```python
EduFactor = df[[
      'TotalFertilityRate',
      'BelowSecondary',
      'Secondary',
      'PostSecondary(NontoTertiary)',
      'Diploma&ProfessionalQualification',
      'University']]

EduFactor = EduFactor.dropna()

EduFactor
```

```python
MarFactor = df[[
      'TotalFertilityRate',
      'FemaleGeneralMarriageRate',
      'M.15to19Years',
      'M.20to24Years',
      'M.25to29Years',
      'M.30to34Years',
      'M.35to39Years',
      'M.40to44Years',
      'M.45to49Years',
      'M.50to54Years',
      'M.55to59Years',
      'M.60to64Years',
      'M.65Years&Over']]


MarFactor = MarFactor.dropna()
MarFactor = MarFactor.replace('na', 0)
MarFactor = MarFactor.replace(',', '', regex=True)
MarFactor
```

```python
CoLFactor = df[[
      'TotalFertilityRate',
      'AllItems',
      'Food',
      'Clothing&Footwear',
      'Housing&Utilities',
      'HouseholdDurables&Services',
      'HouseholdServices&Supplies',
      'HealthCare',
      'Transport',
      'PrivateTransport',
      'PublicTransport',
      'OtherTransportServices',
      'Communication',
      'Recreation&Culture',
      'Education',
      'MiscellaneousGoods&Services',
      'PersonalCare',
      'AlcoholicDrinks&Tobacco',
      'PersonalEffects',
      'AllItemsLessImputedRentalsOnOwnertoOccupiedAccommodation',
      'AllItemsLessAccommodation']]

CoLFactor = CoLFactor.dropna()
CoLFactor = CoLFactor.replace('na', 0)
CoLFactor = CoLFactor.replace(',', '', regex=True)
CoLFactor
```

```python
FerFactor = df[[
      'TotalFertilityRate',
      'F.15to19Years',
      'F.20to24Years',
      'F.25to29Years',
      'F.30to34Years',
      'F.35to39Years',
      'F.40to44Years',
      'F.45to49Years']]


FerFactor = FerFactor.dropna()
FerFactor = FerFactor.replace('na', 0)
FerFactor = FerFactor.replace(',', '', regex=True)
FerFactor
```

```python
#Change all the columns to float64
CoLFactor = CoLFactor.astype(float).round(2)

#Change all the columns to float64
EduFactor = EduFactor.astype(float).round(2)

#Change all the columns to float64
MarFactor = MarFactor.astype(float).round(2)

#Change all the columns to float64
FerFactor = FerFactor.astype(float).round(2)
```

```python
#Change all the 0s to mean value
EduFactor = EduFactor.replace(0, EduFactor.mean())
CoLFactor = CoLFactor.replace(0, CoLFactor.mean())
MarFactor = MarFactor.replace(0, CoLFactor.mean())
FerFactor = FerFactor.replace(0, CoLFactor.mean())
```

### Multi-Var Plots

```python
FerFactor_merge = FerFactor.reset_index()
FerFactor_merge = FerFactor_merge.drop(columns=['TotalFertilityRate'])

FerFactor_merge = pd.melt(FerFactor_merge, id_vars=['Years'], var_name='AgeGroup', value_name='FertilityRate')

sb.relplot(data=FerFactor_merge, x="Years", y="FertilityRate", hue="AgeGroup", kind="scatter", height=6, aspect=1.5)

print("")
```

```python
MarFactor_merge = MarFactor.reset_index()
MarFactor_merge = MarFactor_merge.iloc[:33,:]
MarFactor_merge = MarFactor_merge.drop(columns=['TotalFertilityRate'])
MarFactor_merge = MarFactor_merge.drop(columns=['FemaleGeneralMarriageRate'])

MarFactor_merge = pd.melt(MarFactor_merge, id_vars=['Years'], var_name='AgeGroup', value_name='Marriages per 1000 females')

fig, ax1 = plt.subplots(figsize=(12,6))
sb.lineplot(data = EduFactor['TotalFertilityRate'], marker='o', sort = False, ax=ax1)
ax2 = ax1.twinx()

sb.scatterplot(data=MarFactor_merge, x="Years", y="Marriages per 1000 females", hue="AgeGroup", ax=ax2)

ax1.set_ylabel("")
ax2.set_ylabel("")
plt.title("(Line) Total Fertility Rate                   (Dot) Marriages per 1000 females: ")
ax2.legend(bbox_to_anchor=(1.02, 1.1), loc='upper left', borderaxespad=3)
print("")
```

```python
EduFactor_new = EduFactor
EduFactor_new = EduFactor_new.reset_index()
EduFactor_new = EduFactor_new.iloc[::-1].reset_index(drop=True)

def drawGraph(myCol):
    fig, ax1 = plt.subplots(figsize=(12,6))

    sb.lineplot(data = EduFactor_new['TotalFertilityRate'], marker='o', sort = False, ax=ax1)
    ax2 = ax1.twinx()
    sb.barplot(data = EduFactor_new, x='Years', y=myCol, palette="husl", alpha=0.5, ax=ax2)


    ax1.set_ylabel("")
    ax2.set_ylabel("")
    ax1.set_xticklabels(EduFactor_new['Years'], rotation=90)
    plt.title("(Line) Total Fertility Rate                   (Bar) Females with education: " + myCol)
    print("")
    
drawGraph("BelowSecondary")
```
