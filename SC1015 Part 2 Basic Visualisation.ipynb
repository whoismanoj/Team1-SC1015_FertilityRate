## Part 2: Basic Visualisation

---------------------------
Visualisation is an essential tool for data analysis, as it allows us to communicate insights and patterns that may not be apparent from raw data. In this section, we will explore and compare 4 different data sets using basic visualisation techniques (i.e Boxplot, violin plot, graph plot)

```python
#Basic Libraries
import pandas as pd
import numpy as np

import seaborn as sb
import matplotlib.pyplot as plt 
sb.set() # set the default Seaborn style for graphics
```

### Exploring the Fertility Data (Per 1000 females)

Based on the graphs presented below, it is evident that there has been a notable decrease in the birth rate per 1000 females. For instance, females aged between 25 to 29 years old had a birth rate of around 304.9 per 1000 females in 1961, whereas the corresponding birth rate in 2022 for this age group is 49.1 per 1000 females.

```python
fer_data = pd.read_pickle('FerFactor.pickle')
fer_data
```

```python
fer_data.describe().round(2)
```

```python
# plot the data as a line plot
fer_plot = fer_data.plot(y=['TotalFertilityRate',
                'F.15to19Years',
                'F.20to24Years',
                'F.25to29Years',
                'F.30to34Years',
                'F.35to39Years',
                'F.40to44Years',
                'F.45to49Years'])

# Get the legend object
legend = fer_plot.legend()

# Set the position of the legend outside the plot
legend.set_bbox_to_anchor((1.5, 1))

# display the plot
plt.show()
```

```python
TotalFertilityRate = pd.DataFrame(fer_data['TotalFertilityRate'])
```

```python
f, axes = plt.subplots(1, 3, figsize=(24, 8))

#BelowSecondary
sb.boxplot(data = TotalFertilityRate, orient = "h",ax = axes[0])
sb.histplot(data = TotalFertilityRate,ax = axes[1])
sb.violinplot(data = TotalFertilityRate, orient = "h",ax = axes[2])
```

### Exploring the Education Data

In the dataset under analysis, we have explored the education of females, and observed a consistent increase in the number of females attaining degrees over the years. Our analysis will focus on examining the relationship between the highest qualification attained by females in Singapore and the corresponding birth rate.

```python
edu_data = pd.read_pickle('EduFactor.pickle')
edu_data.head()
```

```python
edu_data.describe().round(2)
```

```python
# plot the data as a line plot
edu_plot = edu_data.plot(y=['BelowSecondary',
                 'Secondary',
                 'PostSecondary(NontoTertiary)',
                 'Diploma&ProfessionalQualification',
                 'University'])
# Get the legend object
legend = edu_plot.legend()

# Set the position of the legend outside the plot
legend.set_bbox_to_anchor((1.05, 1))

# display the plot
plt.show()
```

```python
BelowSecondary = pd.DataFrame(edu_data['BelowSecondary'])
Secondary = pd.DataFrame(edu_data['Secondary'])
PostSecondary = pd.DataFrame({
    'PostSecondary(NontoTertiary)': edu_data['PostSecondary(NontoTertiary)'],
    'Diploma&ProfessionalQualification': edu_data['Diploma&ProfessionalQualification']
})
University = pd.DataFrame(edu_data['University'])
```

```python
f, axes = plt.subplots(4, 3, figsize=(24, 12))

#BelowSecondary
sb.boxplot(data = BelowSecondary, orient = "h",ax = axes[0,0])
sb.histplot(data = BelowSecondary,ax = axes[0,1])
sb.violinplot(data = BelowSecondary, orient = "h",ax = axes[0,2])

#Secondary
sb.boxplot(data = Secondary, orient = "h",ax = axes[1,0])
sb.histplot(data = Secondary,ax = axes[1,1])
sb.violinplot(data = Secondary, orient = "h",ax = axes[1,2])

#PostSecondary
sb.boxplot(data = PostSecondary, orient = "h",ax = axes[2,0])
sb.histplot(data = PostSecondary,ax = axes[2,1])
sb.violinplot(data = PostSecondary, orient = "h",ax = axes[2,2])

#University
sb.boxplot(data = University, orient = "h",ax = axes[3,0])
sb.histplot(data = University,ax = axes[3,1])
sb.violinplot(data = University, orient = "h",ax = axes[3,2])
```

## Exploring the Cost of living Data

In this dataset, we study the cost of living in Singapore and see the trends in the past years. We learned that the cost of living is increasing every year including our education cost, which could be one of the reason for lower fertility rate as it is more expensive to provide kids education. 

```python
CoL_data = pd.read_pickle('CoLFactor.pickle')
CoL_data
```

```python
CoL_data.describe().round(2)
```

```python
# plot the data as a line plot
CoL_plot = CoL_data.plot(y=['AllItems',
                 'Food',
                 'Clothing&Footwear',
                 'Housing&Utilities',
                 'HouseholdDurables&Services',
                'HealthCare',
                'Transport',
                'Communication',
                'Recreation&Culture',
                'Education',
                 'MiscellaneousGoods&Services'])

# Get the legend object
legend = CoL_plot.legend()

# Set the position of the legend outside the plot
legend.set_bbox_to_anchor((1.05, 1))

# display the plot
plt.show()
```

```python
All = pd.DataFrame(CoL_data['AllItems'])

Fertility = pd.DataFrame(CoL_data['TotalFertilityRate'])
Food = pd.DataFrame(CoL_data['Food'])
Clothing = pd.DataFrame(CoL_data['Clothing&Footwear'])
Household = pd.DataFrame({
    'Housing&Utilities': CoL_data['Housing&Utilities'],
    'HouseholdDurables&Services': CoL_data['HouseholdDurables&Services'],
    'HouseholdServices&Supplies' : CoL_data['HouseholdServices&Supplies']
})
Healthcare = pd.DataFrame(CoL_data['HealthCare'])
Transport = pd.DataFrame(CoL_data['Transport'])
Communication = pd.DataFrame(CoL_data['Communication'])
RecreationCulture = pd.DataFrame(CoL_data['Recreation&Culture'])
Education = pd.DataFrame(CoL_data['Education'])
```

```python
f, axes = plt.subplots(8, 3, figsize=(24, 12))

#Food
sb.boxplot(data = Food, orient = "h",ax = axes[0,0])
sb.histplot(data = Food,ax = axes[0,1])
sb.violinplot(data = Food, orient = "h",ax = axes[0,2])

#Clothing
sb.boxplot(data = Clothing, orient = "h",ax = axes[1,0])
sb.histplot(data = Clothing,ax = axes[1,1])
sb.violinplot(data = Clothing, orient = "h",ax = axes[1,2])

#Household
sb.boxplot(data = Household, orient = "h",ax = axes[2,0])
sb.histplot(data = Household,ax = axes[2,1])
sb.violinplot(data = Household, orient = "h",ax = axes[2,2])

#Healthcare
sb.boxplot(data = Healthcare, orient = "h",ax = axes[3,0])
sb.histplot(data = Healthcare,ax = axes[3,1])
sb.violinplot(data = Healthcare, orient = "h",ax = axes[3,2])

#Transport
sb.boxplot(data = Transport, orient = "h",ax = axes[4,0])
sb.histplot(data = Transport,ax = axes[4,1])
sb.violinplot(data = Transport, orient = "h",ax = axes[4,2])

#Communication
sb.boxplot(data = Communication, orient = "h",ax = axes[5,0])
sb.histplot(data = Communication,ax = axes[5,1])
sb.violinplot(data = Communication, orient = "h",ax = axes[5,2])

#RecreationCulture
sb.boxplot(data = RecreationCulture, orient = "h",ax = axes[6,0])
sb.histplot(data = RecreationCulture,ax = axes[6,1])
sb.violinplot(data = RecreationCulture, orient = "h",ax = axes[6,2])

#Education
sb.boxplot(data = Education, orient = "h",ax = axes[7,0])
sb.histplot(data = Education,ax = axes[7,1])
sb.violinplot(data = Education, orient = "h",ax = axes[7,2])
```
