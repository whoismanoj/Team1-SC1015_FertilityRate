```python
#Basic Libraries
import pandas as pd
import numpy as np

import seaborn as sb
import matplotlib.pyplot as plt 
sb.set() # set the default Seaborn style for graphics
```

### Exploring the education Data

```python
edu_data = pd.read_pickle('EduFactor.pickle')
edu_data
```

```python
edu_data.describe().round(2)
```

```python
# plot the data as a line plot
edu_data.plot(y=['BelowSecondary',
                 'Secondary',
                 'PostSecondary(NontoTertiary)',
                 'Diploma&ProfessionalQualification',
                 'University'])

# display the plot
plt.show()
```

```python

# plot the data as a line plot
edu_data.plot(y=['TotalFertilityRate'])

# display the plot
plt.show()
```

```python
Fertility = pd.DataFrame(edu_data['TotalFertilityRate'])
BelowSecondary = pd.DataFrame(edu_data['BelowSecondary'])
Secondary = pd.DataFrame(edu_data['Secondary'])
PostSecondary = pd.DataFrame({
    'PostSecondary(NontoTertiary)': edu_data['PostSecondary(NontoTertiary)'],
    'Diploma&ProfessionalQualification': edu_data['Diploma&ProfessionalQualification']
})
University = pd.DataFrame(edu_data['University'])
```

```python
f, axes = plt.subplots(5, 3, figsize=(24, 12))

#Fertility
sb.boxplot(data = Fertility, orient = "h",ax = axes[0,0])
sb.histplot(data = Fertility,ax = axes[0,1])
sb.violinplot(data = Fertility, orient = "h",ax = axes[0,2])

#BelowSecondary
sb.boxplot(data = BelowSecondary, orient = "h",ax = axes[1,0])
sb.histplot(data = BelowSecondary,ax = axes[1,1])
sb.violinplot(data = BelowSecondary, orient = "h",ax = axes[1,2])

#Secondary
sb.boxplot(data = Secondary, orient = "h",ax = axes[2,0])
sb.histplot(data = Secondary,ax = axes[2,1])
sb.violinplot(data = Secondary, orient = "h",ax = axes[2,2])

#PostSecondary
sb.boxplot(data = PostSecondary, orient = "h",ax = axes[3,0])
sb.histplot(data = PostSecondary,ax = axes[3,1])
sb.violinplot(data = PostSecondary, orient = "h",ax = axes[3,2])

#University
sb.boxplot(data = University, orient = "h",ax = axes[4,0])
sb.histplot(data = University,ax = axes[4,1])
sb.violinplot(data = University, orient = "h",ax = axes[4,2])
```

```python
# Create a joint dataframe by concatenating the two variables
jointDF = pd.concat([Fertility, University], axis = 1)
jointDF
```

```python
sb.jointplot(data = jointDF, x = "University", y = "TotalFertilityRate", sizes=(10, 100))
```
